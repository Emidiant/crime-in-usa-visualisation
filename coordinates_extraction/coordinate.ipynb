# Преобразования координат

```python
import pandas as pd
from ast import literal_eval

import json
```

```python
def centroid(vertexes):
    _x_list = [vertex[0] for vertex in vertexes]
    _y_list = [vertex[1] for vertex in vertexes]
    _len = len(vertexes)
    _x = sum(_x_list) / _len
    _y = sum(_y_list) / _len
    return _x, _y
```

```python
coordinates = pd.read_csv("state_coordinates/csv/coordinates.csv")
# coordinates.set_index("state", inplace=True)
coordinates
```

```python
def coord_transform(x):
    coord = literal_eval(x["coordinates"])
    if x["type"] == "polygon":
        raw_coord_df = pd.DataFrame(coord, columns =['x', 'y'])
        first_coord_df = raw_coord_df.copy()
        if x["state"] != "Alaska" and x["state"] != "Hawaii":
            first_coord_df["x"] = 13 * (first_coord_df["x"] + 129)
            first_coord_df["y"] = 13 * (first_coord_df["y"] + 54)
        elif x["state"] == "Alaska":
            first_coord_df["x"] = 6 * (first_coord_df["x"] + 200)
            first_coord_df["y"] = 6 * (first_coord_df["y"] + 300)
        elif x["state"] == "Hawaii":
            first_coord_df["x"] = 5 * (first_coord_df["x"] + 180)
            first_coord_df["y"] = 5 * (first_coord_df["y"] + 50)

        first_coordinates = first_coord_df.values.tolist()
        min_max_df = pd.DataFrame(first_coordinates, columns =['x', 'y'])
        print(f"state: {x['state']}, X in [{round(min_max_df['x'].min(),3)}, {round(min_max_df['x'].max(),3)}], Y in [{round(min_max_df['y'].min(),3)}, {round(min_max_df['y'].max(), 3)}] ")
        return first_coordinates
    else:
        new_coords = []
        for poly in literal_eval(x["coordinates"]):
            new_coords.append(coord_transform({"state": x['state'], "coordinates": str(poly), "type": "polygon"}))
        return new_coords
```

```python
coordinates["new_coordinates"] = coordinates[['coordinates', 'type', 'state']].apply(lambda x: coord_transform(x), axis=1)
coordinates
```

```python
coordinates[["state", "type", "new_coordinates"]].to_csv("state_coordinates/csv/polygon.csv", index=False)
```

```python
part_crime = crime[["year", "murder", "state", "total"]].drop_duplicates()
```
